```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time, random, pandas
```

```python
driver = webdriver.Chrome(r'C:\Users\sivakumar\Desktop\chromedriver.exe')
driver.get(r'https://www.nytimes.com/2018/06/05/technology/facebook-device-partnerships-china.html#commentsContainer')
```

```python
driver.find_element_by_class_name('Tabs-nav--1EAdz').find_elements_by_tag_name('a')[2].click()
length = 218/25
```

```python
try:
    for i in range(length) :
        driver.find_element_by_class_name('CommentList-viewMore--2BjZl').click()
except:
    print("click ended")
comments = driver.find_elements_by_class_name('Comment-commentText--1826c')
```

```python
comment_text =[]
for comment in comments:
    print(comment.text) 
    comment_text.append(comment.text)
    print("----------------------------")
```

```python
results = pandas.DataFrame(comment_text)
results.to_csv('comments.csv')
```
